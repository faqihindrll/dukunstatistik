```python
def fetch_data_for_year(start_year, queries):
    params = {
        "engine": "google_trends",
        "q": ", ".join(queries),  # Menggabungkan query menjadi satu string
        "data_type": "GEO_MAP",
        "date": f"{start_year}-01-01 {start_year}-12-31",
        "geo": "ID-JT",
        "region": "CITY",
        "api_key": api_faqih,  # Ganti dengan API key Anda
    }

    search = GoogleSearch(params)
    results_test = search.get_dict()

    # Dictionary untuk menyimpan hasil
    data = {query: [] for query in queries}
    locations = []

    for region in results_test.get('compared_breakdown_by_region', []):
        if region['location'] in ['Slawi', 'Tegal','Adiwerna']:
            locations.append(region['location'])
            for value in region['values']:
                if value['extracted_value'] != '':
                    data[value['query']].append(value['extracted_value'])

    # Membuat DataFrame dari hasil yang didapat
    df = pd.DataFrame({
        'Kota': locations,
    })
    for query in queries:
        df[query] = data[query]

    return df
```

```python
def save_data_for_years(start_year, end_year, queries, batch_size=5):
    num_batches = (len(queries) + batch_size - 1) // batch_size
    for year in range(start_year, end_year + 1):
        for i in range(num_batches):
            batch_queries = queries[i*batch_size:(i+1)*batch_size]
            df = fetch_data_for_year(year, batch_queries)
            if df.empty:
                print(f"Tidak terdapat data untuk tahun {year} batch {i+1}.")
            else:
                df.to_excel(f"kemiskinan_{year}_batch_{i+1}.xlsx", index=False)
                print(f"Data untuk tahun {year} batch {i+1} berhasil disimpan.")
```

## 1 Query (Pengangguran)

```python
import pandas as pd

def fetch_data_for_year(start_year, query, api_key):
    params = {
        "engine": "google_trends",
        "q": query,  # Hanya satu query
        "data_type": "GEO_MAP_0",
        "date": f"{start_year}-01-01 {start_year}-12-31",
        "geo": "ID-JT",  # Area Jawa Tengah
        "region": "CITY",
        "api_key": api_key,  # API key
    }

    search = GoogleSearch(params)
    results_test = search.get_dict()

    # Dictionary untuk menyimpan hasil
    data = {query: []}
    locations = []

    for region in results_test.get('interest_by_region', []):
        location = region.get('location', '')
        extracted_value = region.get('extracted_value', '')

        if location in ['Slawi', 'Tegal', 'Adiwerna']:  # Filter lokasi
            locations.append(location)
            data[query].append(extracted_value)

    # Membuat DataFrame dari hasil yang didapat
    df = pd.DataFrame({
        'Kota': locations,
        query: data[query]
    })

    return df

def scrape_trends_and_save(start_year, end_year, query, api_key, output_file):
    all_data = pd.DataFrame()

    # Looping dari tahun 2010 sampai 2024
    for year in range(start_year, end_year + 1):
        print(f"Fetching data for {year}...")
        yearly_data = fetch_data_for_year(year, query, api_key)
        
        # Menambahkan kolom tahun
        yearly_data['Tahun'] = year

        # Menggabungkan data tahun tersebut ke dalam DataFrame utama
        all_data = pd.concat([all_data, yearly_data], ignore_index=True)

    # Menyimpan hasil akhir ke file Excel
    all_data.to_excel(output_file, index=False)
    print(f"Data saved to {output_file}")
```

```python
scrape_trends_and_save(2010, 2024, 'lowongan kerja', api_faqih,"gti_lowongan_kerja.xlsx")
```

## Compile (Many Queries Only)

```python
import pandas as pd
import os

def compile_files(start_year, end_year, batches):
    compiled_df = pd.DataFrame()

    for year in range(start_year, end_year + 1):
        for batch in batches:
            file_name = f"kemiskinan_{year}_batch_{batch}.xlsx"
            if os.path.exists(file_name):
                df = pd.read_excel(file_name)
                df['Year'] = year  # Tambahkan kolom tahun
                df['Batch'] = batch  # Tambahkan kolom batch
                compiled_df = pd.concat([compiled_df, df], ignore_index=True)
            else:
                print(f"File {file_name} tidak ditemukan.")

    compiled_df.to_excel(f"compiled_data_{start_year}_{end_year}.xlsx", index=False)
    print("Semua file berhasil dikompilasi menjadi compiled_data.xlsx.")
```

```python
start_year = 2024
end_year = 2024
batches = [1, 2, 3, 4]  # Sesuaikan dengan jumlah batch yang Anda miliki
compile_files(start_year, end_year, batches)
```

## Imputasi KNN

```python
from sklearn.impute import KNNImputer
import pandas as pd
# Load the data from the specified sheet "compile"
df = pd.read_excel("gti_lowongan_kerja.xlsx", sheet_name="Data")

# Kolom yang akan diimputasi
columns_to_impute = df.columns.difference(['Tahun'])

n_neighbors = 2
# Initialize the KNNImputer
imputer = KNNImputer(n_neighbors=n_neighbors)

# Apply KNN imputation only on the selected columns and only where there are missing values
df[columns_to_impute] = imputer.fit_transform(df[columns_to_impute])
```
